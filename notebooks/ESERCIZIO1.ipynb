```python

import pandas as pd
import os
Matrice=pd.read_excel('../data/PCA-Data-Esercise.xlsx', sheet_name=0) # in principio gli dici il foglio
Matrice=Matrice.dropna(how='all')
Matrice=Matrice.dropna(axis=1,how='all') #rimuove nan da colonne =1 e righe=0
print(Matrice)
```

```python
rounded_corr_matrix = Matrice.corr().round(4)
print(rounded_corr_matrix)
```

```python
import numpy as np

# Defining a square matrix
matrix = np.array(rounded_corr_matrix)

# Eigenvalues and eigenvectors for diagonalization
eigenvalues, eigenvectors = np.linalg.eig(matrix)

# Print the results
print("Eigenvalues:", eigenvalues.round(4))
print("Eigenvectors:\n", eigenvectors)

```

```python
somma_eigval=sum(eigenvalues) # traccia
array_weight = np.zeros(7)
a=0
for i in eigenvalues:
    
    percentuale=i/somma_eigval
    array_weight[a]=percentuale
    a=a+1

print(array_weight)
```
