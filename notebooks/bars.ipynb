```python
import matplotlib.pyplot as plt
import numpy as np

# Dati
labels = ['Accuracy', 'Sensitivity']
rf_means = [0.753, 0.785]
rf_std = [0.189, 0.243]
svm_means = [0.721, 0.646]
svm_std = [0.198, 0.317]

x = np.arange(len(labels)) 
width = 0.35 

fig, ax = plt.subplots(figsize=(6, 4), dpi=200)

# Barre Random Forest
ax.bar(x - width/2, rf_means, width, yerr=rf_std, label='Random Forest', 
       capsize=5, color='skyblue', edgecolor='black')

# Barre SVM
ax.bar(x + width/2, svm_means, width, yerr=svm_std, label='SVM', 
       capsize=5, color='orange', edgecolor='black', alpha=0.8)

# Dettagli minimi
ax.set_title('Confronto Modelli')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.1)
ax.legend()

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
